```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
data = pd.read_excel("GRN_source_Target_TFS.xls")
```

```python
data.head()
```

```python
source = data['TF'].values.tolist()
```

```python
source[0:5]
```

```python
dest = data['Target'].values.tolist()
```

```python
dest[0:5]
```

```python
def merge(list1, list2):
      
    merged_list = [(list1[i], list2[i]) for i in range(0, len(list1))]
    return merged_list

src_dest = merge(source,dest)
```

```python
src_dest[0:5]
```

```python
src_dest[0]
```

```python
import networkx as nx 
```

```python
G = nx.DiGraph()
```

```python
G.add_edges_from(src_dest)
```

```python
plt.figure(figsize =(10,5))
nx.draw(G)
```

```python
nx.is_directed_acyclic_graph(G) #cycles exist
```

```python
nx.DiGraph.order(G)
```

```python
nx.DiGraph.size(G)
```

```python
# Removing cycles 
# Logic - while DAG is False find a cycle from G and remove that cycle until DAG is true. 
while nx.is_directed_acyclic_graph(G) == False : 
    cycle = nx.find_cycle(G)
    #print(cycle)
    G.remove_edges_from(cycle)
```

```python
nx.is_directed_acyclic_graph(G) #no cycle exist. Directed Acyclic graph.
```

```python
nx.DiGraph.size(G)
```

```python
#nx.number_of_isolates(G)
```

```python
plt.figure(figsize =(12, 5))
nx.draw(G)
```

```python
score = pd.read_excel("GRN_source_target_network_analyze.xls")
```

```python
score.head()
```

```python
cent = score['ClosenessCentrality'].values.tolist()
```

```python
cent[0:5]
```

```python
name= score['name'].values.tolist()
name[0:5]
```

```python
centrality = {}
for key in name:
    for value in cent:
        centrality[key] = value
        cent.remove(value)
        break  
```

```python
centrality['ABL1']
```

```python
isinstance(centrality, dict)
nx.set_node_attributes(G,centrality,name="Centrality")
```

```python
from pgmpy.factors.discrete import TabularCPD
from pgmpy.models import BayesianModel
```

```python
model = BayesianModel([('TP53','AR'),
                      ('MYC','AR'),
                      ('AR','ESR1')])
```

```python
tp = TabularCPD(
    variable= 'TP53',
    variable_card= 2, 
    values = [[0.46456693],[0.0001]]
)
```

```python
myc = TabularCPD(
    variable= 'MYC',
    variable_card= 2, 
    values = [[0.35014837],[0.01]]
)
```

```python
esr = TabularCPD(
    variable= 'MYC',
    variable_card= 2, 
    values = [[0.4469697,0.01],[0.01,.44]],
    evidence= ['AR'],
    evidence_card=[2]
)
```

```python
ar = TabularCPD(
    variable= 'AR',
    variable_card= 2, 
    values = [[0.40410959,0.01,0.464,0.35],[0.01,0.404,0.464,0.35]],
    evidence = ['TP53','MYC'],
    evidence_card = [2,2]
)
```

```python
model.add_cpds(tp,myc,esr,ar)
```

```python
model.get_cpds()
```

```python
model.active_trail_nodes('TP53')
```

```python
model.active_trail_nodes('AR')
```

```python
model.local_independencies('TP53')
```

```python
model.local_independencies('AR') #nothing is independent. 
```

```python
model.get_independencies()
```
